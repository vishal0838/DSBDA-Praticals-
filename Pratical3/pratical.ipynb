```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Titanic-Dataset.csv")
df.head()
```

```python
summary_stats = df.groupby('Sex')['Age'].agg(['mean', 'median', 'min', 'max', 'std'])

df['Sex_Numeric'] = df['Sex'].map({'male':0, 'female': 1})
numeric_response = df['Sex_Numeric'].tolist()

print("----- Summary Statictics of Age by Sex ------")
display(summary_stats)

print("\n ------ Numeric List for Categorical 'Sex' -------")
print(numeric_response)
```

```python
for p_class in sorted(df['Pclass'].unique()): 
    print(f"\n ------- Statistical Details for Pclass {p_class} -----")

    class_df = df[df['Pclass'] == p_class]

    display(class_df.describe())
```

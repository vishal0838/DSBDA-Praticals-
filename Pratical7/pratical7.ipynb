```python
pos = pos_tag(tokens)
print("POS Tags:\n", pos)
```

```python
stop_words = set(stopwords.words('english'))
filtered = [word for word in tokens if word.lower() not in stop_words]

print("After Stopword Removal:\n", filtered)
```

```python
ps = PorterStemmer()
stemmed = [ps.stem(word) for word in filtered]

print("Stemmed Words:\n", stemmed)
```

```python
lemmatizer = WordNetLemmatizer()
lemmatized = [lemmatizer.lemmatize(word) for word in filtered]

print("Lemmatized Words:\n", lemmatized)
```

```python
documents = [text]

vectorizer = TfidfVectorizer()
tfidf = vectorizer.fit_transform(documents)

print("Words:\n", vectorizer.get_feature_names_out())
print("\nTF-IDF Matrix:\n", tfidf.toarray())
```

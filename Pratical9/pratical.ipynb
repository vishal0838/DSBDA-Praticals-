```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset("titanic")
df.head()
```

```python
df = df.dropna(subset=["age"])
```

```python
sns.boxplot(x="sex", y="age", hue="survived", data=df)

plt.title("Age Distribution by Gender and Survival")
plt.xlabel("Gender")
plt.ylabel("Age")
plt.show()
```

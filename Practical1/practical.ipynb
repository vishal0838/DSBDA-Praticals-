```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
```


# Step 2: Load Dataset
# Download dataset from:
# https://www.kaggle.com/datasets/yasserh/titanic-dataset

```python
df = pd.read_csv("Titanic-Dataset.csv")
```

```python
# Step 3: Display First 5 Rows
print("\nFirst 5 Rows of Dataset")
print(df.head())
```

```python
# Step 4: Dataset Information
print("\nDataset Information")
print(df.info())
```

```python
# Step 5: Dataset Description
print("\nStatistical Description")
print(df.describe())
```

```python
# Step 6: Check Missing Values
print("\nMissing Values")
print(df.isnull().sum())
```

```python
# Step 7: Check Dimensions of Dataset
print("\nShape of Dataset")
print(df.shape)
```

```python
# Step 8: Display Data Types
print("\nData Types")
print(df.dtypes)
```

```python
# Step 9: Fill Missing Values
df['Age'] = df['Age'].fillna(df['Age'].mean())
```

```python
df['Sex'] = df['Sex'].map({'male':0, 'female':1})
```

```python
print("Dataset After Categorical Value Replaced\n")
df.head()
```

```python
# Step 11: Data Normalization
scaler = MinMaxScaler()

df[['Age', 'Fare']] = scaler.fit_transform(df[['Age', 'Fare']])

print("\nNormalized Data")
print(df[['Age', 'Fare']].head())
```

```python
# Step 12: Final Dataset
print("\nFinal Processed Dataset")
print(df.head())
```
